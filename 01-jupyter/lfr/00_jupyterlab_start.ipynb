## 第三步：导入工具库

科研 Notebook 通常把 `import` 集中放在前面。`numpy` 用于数组计算，`matplotlib` 用于画图。先运行这一格；如果没有红色报错，就表示导入成功。

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(42)
````

## 第四步：生成数组并查看它

数组可以理解为一组按顺序存放的数。运行后观察：`shape` 表示数组有几个元素。

````python
x = np.arange(0, 10)
print('数组 x：', x)
print('x 的形状：', x.shape)
print('x 中第 4 个元素：', x[3])
````

## 第五步：画第一张图

图会直接出现在代码格下方。试着把 `2` 改成 `3`，再次运行，比较曲线变化。

````python
x = np.linspace(-3, 3, 100)
y = x ** 3

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.title('My First Figure')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
````
